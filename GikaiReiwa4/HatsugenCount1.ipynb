```python
fnames = []
with open('filelist.txt', 'r', encoding= "utf-8-sig") as f:
    for line in f:
        fnames.append(line.strip())
print(fnames)
```

```python
len(fnames)
```

```python
import re
import collections
iin2=[]

for fname in fnames:
    print(fname)
    with open (fname,'r', encoding= "utf-8-sig") as g:
        iin2=[]
        for line in g:
            #iin=re.findall(r'\○(.*)\u3000',line )
            iin=re.search(r'○(.+)\u3000',line)
            if iin is not None: iin2.append(iin.group(1))
    c = collections.Counter(iin2)
    
    print(c)
    print("\n")
```
